## Mostrar las primeras 10 observaciones

```python
# Cargamos el dataset
datos = pd.read_csv("https://raw.githubusercontent.com/rpizarrog/Analisis-Inteligente-de-datos/main/datos/datos.FIFA.limpios.csv", encoding='latin1')
datos.head(10)
```

## Describir estructura de los datos

```python
datos.to_json
```

## Resolver e interpretar cada una de las 10 preguntas

### ¿Qué jugadores ganan más dinero, el top 5?

```python
valor = datos.sort_values(['Valor'], ascending=False)
valor[["Name", "Valor"]].head()
```

### ¿Cuántos jugadores de fútbol son izquierdos?

```python
frecuencia_pie = (datos.groupby("Preferred.Foot").agg(frequency=("Preferred.Foot", "count")))
type(frecuencia_pie)
# frecuencia_nac
frecuencia_pie = frecuencia_pie.sort_values(['frequency'], ascending=True)
frecuencia_pie
```

### ¿Cuál es el valor medio (promedio) de estaura en cms. o metros de jugadores de fútbol?

```python
estatura_promedio = round(datos["Estatura"].mean(),2)

print("La estatura promedio de los jugadores es de",estatura_promedio,"m")
```

### ¿Cómo se llaman los jugadores más bajitos en estatura?

```python
estaturas_sorted = datos.sort_values(['Estatura'], ascending=True)
estaturas_sorted.head()[["Name", "Estatura"]]
```

### ¿De qué nacionalidad o país hay más jugadores registrados?

```python
frec_nacionalidad = (datos.groupby("Nationality")).agg(frequency=("Nationality", "count"))
type(frec_nacionalidad)
frec_nacionalidad = frec_nacionalidad.sort_values(['frequency'], ascending=False)
frec_nacionalidad.head(1)
```
