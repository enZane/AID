# Cargar datos
Los datos se encuentran en la URL: https://raw.githubusercontent.com/rpizarrog/Analisis-Inteligente-de-datos/main/datos/datos.FIFA.limpios.csv

```python
datos = pd.read_csv("https://raw.githubusercontent.com/rpizarrog/Analisis-Inteligente-de-datos/main/datos/datos.FIFA.limpios.csv", encoding='latin1')
datos
```

# Estadísticos de variable de interés
## Nacionalidad

```python
nacionalidad = pd.Categorical(datos['Nationality'])
nacionalidad.describe()
```

```python
frecuencia_nac = (datos.groupby("Nationality").agg(frequency=("Nationality", "count")))
type(frecuencia_nac)
# frecuencia_nac
frecuencia_nac = frecuencia_nac.sort_values(['frequency'], ascending=False)
frecuencia_nac_topten = frecuencia_nac.head(10)
frecuencia_nac_topten
```

```python
frecuencia_nac_topten.plot(kind="bar")
```

# Pie preferido

```python
pie_preferido = pd.Categorical(datos['Preferred.Foot'])
pie_preferido.describe()
```

```python
frecuencia_pie = (datos.groupby("Preferred.Foot").agg(frequency=("Preferred.Foot", "count")))
type(frecuencia_pie)
# frecuencia_nac
frecuencia_pie = frecuencia_pie.sort_values(['frequency'], ascending=True)
frecuencia_pie
```

```python
frecuencia_pie.plot(kind="bar")
```

## Edad de jugadores

```python
print("Age")
print("La media aritmética de edad es :",round(datos['Age'].mean(),0), " años")
```

## Agresión

```python
agresivos = datos.sort_values(['Aggression'], ascending=False)
agresivos.head()
```

## Valor

```python
# order datos by Valor picking name and valor
valor = datos.sort_values(['Valor'], ascending=False)

valor[["Name", "Valor"]].head()
```

## Estatura

```python
estatura_promedio = round(datos["Estatura"].mean(),2)

print("La estatura promedio de los jugadores es de",estatura_promedio,"m")
```

```python
estaturas_sorted = datos.sort_values(['Estatura'], ascending=True)
estaturas_sorted.head()["Name"]
```
